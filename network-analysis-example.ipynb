# Network analysis example

```python
# Import libraries
import process.improvement as PI
import pandas as pd
```

```python
# Specify path to data
dataset_url = "https://raw.githubusercontent.com/jimlehner/datasets/main/network-analysis-example-data.csv"
```

```python
# Read csv from URL
def get_data() -> pd.DataFrame:
    return pd.read_csv(dataset_url, index_col=0)

# Get data
df = get_data()
df.head()
```

```python
# Break DataFrame into a list of DataFrames with the same column title of 'Values'
list_of_dfs = [df[[col]].rename(columns={col: 'Values'}) for col in df.columns]
```

```python
# Generate list of labels based on df columns
list_of_labels = df.columns.tolist()
list_of_labels
```

```python
# Perform network_analysis on dataset
results_df = PI.network_analysis(list_of_dfs, 'Values', list_of_labels, rows=2, cols=4, figsize=(15,6), title='')
```
